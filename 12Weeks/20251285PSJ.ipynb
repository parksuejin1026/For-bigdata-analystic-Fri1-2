```python
from matplotlib.lines import lineStyles
import matplotlib.pyplot as plt
# 그래프는 항상 x축과 y축의 값 대응이 되어야함
years = [1950, 1960, 1970, 1980, 1990, 2000, 2010]
gdp = [67.0, 80.0, 267.0, 1686.0, 6505, 11865.3, 22105.3]


plt.plot(years, gdp, color = 'green', marker = 'o', linestyle = 'solid')

plt.title("GDP per capital")

plt.ylabel("dollars")
plt.savefig("gdp_per_capita.png", dpi = 600)
plt.show()
```

```python
x = [x for x in range(-20,20)]
y1 = [2*t for t in x]
y2 = [(t**2)+5 for t in x]
y3 = [-(t**2) - 5 for t in x]
# 빨강색 점선, 녹색 실선과 세모기호, 파랑색 별표와 점선으로 각각의 함수를 표현
plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30]) # x와 y값의 범위
plt.savefig("quadratic2.png", dpi = 600)
plt.show()
```
