```python
import matplotlib.pyplot as plt # Matplotlib 라이브러리를 plt라는 별칭으로 임포트합니다.
import numpy as np
```

```python
# 연도별 GDP 데이터를 시각화합니다.

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] # 연도 데이터 리스트
gdp = [67.0, 80.0, 267.0, 1686.0, 6505, 11865.3, 22105.3] # 해당 연도의 GDP 데이터 리스트

# 꺾은선 그래프 생성: x축은 years, y축은 gdp. 초록색 원형 마커와 실선 사용.
plt.plot(years, gdp, color = 'green', marker = 'o', linestyle = 'solid')

plt.title("GDP per capital") # 그래프 제목 설정
plt.ylabel("dollars") # y축 레이블 설정
plt.savefig("gdp_per_capita.png", dpi = 600) # 그래프를 PNG 파일로 저장 (600 dpi 해상도)
plt.show() # 그래프 표시
```

```python
# 여러 함수를 한 그래프에 다양한 스타일로 표현합니다.

x = [i for i in range(-20,20)] # -20부터 19까지의 정수 리스트 (x축 값)
y1 = [2*t for t in x] # y = 2x 함수
y2 = [(t**2)+5 for t in x] # y = x^2 + 5 함수
y3 = [-(t**2) - 5 for t in x] # y = -x^2 - 5 함수

# 세 개의 함수를 한 그래프에 그립니다.
# y1: 빨간색 점선 ('r--')
# y2: 녹색 세모 마커와 실선 ('g^-')
# y3: 파란색 별표 마커와 짧은 점선 ('b*:')
plt.plot(x, y1, 'r--', x, y2, 'g^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30]) # x축과 y축의 범위 설정: x는 -30~30, y는 -30~30
plt.savefig("quadratic2.png", dpi = 600) # 그래프를 PNG 파일로 저장 (600 dpi 해상도)
plt.show() # 그래프 표시
```

```python
# 하나의 차트에 여러 개의 데이터와 범례를 추가하여 시각화하고 파일로 저장합니다.

x = [i for i in range(20)] # 0부터 19까지의 정수 리스트 (x축 값)
y = [i ** 2 for i in range(20)] # x의 제곱 값 리스트 (y축 값: 2차 함수)
z = [ i ** 3 for i in range(20)] # x의 세제곱 값 리스트 (y축 값: 3차 함수)

# 각 함수를 꺾은선 그래프로 그리고, label을 설정하여 범례에 표시될 이름을 지정합니다.
plt.plot(x, x, label='linear') # y = x (선형 함수)
plt.plot(x, y, label='quadratic') # y = x^2 (2차 함수)
plt.plot(x, z, label='cubic') # y = x^3 (3차 함수)

plt.xlabel('x label') # x축 레이블 설정
plt.ylabel('y label') # y축 레이블 설정
plt.title('My Plot') # 그래프 제목 설정
plt.legend() # 범례 표시 (각 plot의 label을 사용)
plt.savefig("cubic3.png", dpi = 600) # 그래프를 PNG 파일로 저장 (600 dpi 해상도)
plt.show() # 그래프 표시
```

```python
# 연도별 GDP 데이터를 막대형 차트로 시각화합니다.

years = [1950, 1960, 1970, 1980, 1990, 2000, 2010] # 연도 데이터 리스트
gdp = [67.0, 80.0, 267.0, 1686.0, 6505, 11865.3, 22105.3] # 해당 연도의 GDP 데이터 리스트

# 막대형 차트 생성: x축은 years 리스트의 인덱스 (0부터 len(years)-1), y축은 gdp 값.
# range(len(years))는 years 리스트의 길이만큼 0부터 시작하는 정수 범위를 생성합니다.
plt.bar(range(len(years)), gdp)

plt.title("GDP per capita") # 그래프 제목 설정
plt.ylabel("dollars") # y축 레이블 설정

# x축 틱(눈금) 설정: range(len(years)) 위치에 years 리스트의 실제 연도를 표시.
# xticks() 함수는 가로축 눈금의 위치와 해당 눈금에 표시될 레이블을 지정합니다.
plt.xticks(range(len(years)), years)
plt.savefig("capita4.png", dpi = 600) # 그래프를 PNG 파일로 저장 (600 dpi 해상도)
plt.show() # 그래프 표시
```

```python
#여러 나라의 국민소득 추이를 다중 막대형 차트로 그리기
years = [1965, 1975, 1985, 1995, 2005, 2015]
ko = [130, 650, 2450, 11600, 17790, 27250]
jp = [890, 5120, 11500, 42130, 40560, 38760]
ch = [100,200,290, 540, 1760, 7940]

x_range = range(len(years))

plt.bar(x_range, jp, width = 0.25) # 큰값을 젤 먼저 놔둬야 다른 값의 그래프가 없어지지 않음 일본 칠하고 한국 칠하고 중국 칠하는 느낌
plt.bar(x_range, ko, width= 0.25) # first는 일본, second는 한국, third는 중국
plt.bar(x_range, ch, width= 0.25)
plt.xticks(x_range, years)
plt.savefig("bar5.png", dpi = 600)
plt.show()
```

```python
x_range = np.arange(len(years))

plt.bar(x_range + 0.0 , jp, width = 0.25) # x값을 더해서 구분하기 쉽게 만들기
plt.bar(x_range + 0.3, ko, width = 0.25)
plt.bar(x_range + 0.6, ch, width = 0.25)
plt.title('GDP per capita')
plt.xlabel("years")
plt.ylabel("dollars")
plt.xticks(x_range, years)
plt.savefig("arange6.png", dpi = 600)
plt.show()
```

```python
# 산포도 그래프 그리기 scatter() 함수

xData = np.arange(20, 50)
yData = xData + 2 * np.random.randn(30)

plt.scatter(xData, yData)
plt.title('Real Age vs Physical Age')
plt.xlabel('Real Age')
plt.ylabel('Physical Age')

plt.savefig("age.png", dpi = 600)
plt.show()
```

```python
# 원형 비율을 표시하는 파이 차트 만들기

times = [8, 14, 2]
timelabels = ["Sleep", "Study", "Play"]
plt.pie(times,
        labels = timelabels, # times의 값에 대한 설명
        autopct = "%.2f")
plt.savefig("pip8.png", dpi = 600)
plt.show()
```

```python
# 히스토그램 : 주어진 자료를 몇 개의 구간으로 나누고 각 구간의 도수를 조사하여 나타낸 막대 그래프
books = [ 1, 6, 2, 3, 1, 2, 0 ,2]
plt.hist(books, bins = 6)

plt.xlabel("books")
plt.ylabel("frequency")
plt.savefig("books9.png", dpi = 600)
plt.show()
```

```python
# box plot 데이터의 최대, 최소, 중간값, 사분위 값을 출력해줌
random_data = np.random.randn(100)

plt.boxplot(random_data)
plt.savefig("box10.png", dpi = 600)
plt.show()
```
