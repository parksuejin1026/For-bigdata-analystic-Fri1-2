```python
# 텍스트 처리
s = 'This movie was terrible! The acting was poor and the story was boring.'
print(f's = {s}')
print(f'\n글자수 = {len(s)}')
words = s.split()
print(f'단어들의 리스트 = {words}')
print(f'단어 수 = {len(words)}')
print(f'평균 단어 길이 = {sum(len(word) for word in words)/len(words)}')
```

```python
# 불용어 제거 실습
stop_words = ['is', 'and', 'the', 'a', 'an', 'in', 'of', 'to']

text = 'This is example of removing stop words from a string'
words = text.split()
fitered_sentence = [word for word in words if word.lower() not in stop_words]

print(f"입력 = {text}")
print(f"출력 = {" ".join(fitered_sentence)}")
```
