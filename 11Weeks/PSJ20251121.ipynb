```python
# 라이브러리 호출 코드 모음
import matplotlib.pyplot as plt
from textblob import TextBlob # 영화평 분석
from wordcloud import WordCloud
from wordcloud import STOPWORDS # 불용어 처리해주는 라이브러리
from PIL import Image
import numpy as np
```

```python

# 영화평 분석

review = "This movie was terrible The acting was poor and the story was boring."
sentiment = TextBlob(review).sentiment.polarity
print(f"입력 = {review}\n")

if sentiment > 0:
  print("결과 : 긍정적")
elif sentiment == 0:
  print("결과 : 중립")
else:
  print("결과 : 부정적")
```

```python
# 메모장에 적은 파일 모두 읽기
f = open('/wikiAI.txt','r',encoding='UTF-8')
text = f.read()
print(text)
```

```python
# 그림으로 출력
wc = WordCloud()
wc.generate(text)
plt.imshow(wc)
plt.show()
```

```python
# 그림으로 출력
wc = WordCloud(width=1500, height=1000)
wc.generate(text)
plt.imshow(wc)
plt.show()
```

```python
불용어  = STOPWORDS.union({'goals', 'field', 'human', 'see'})
wc = WordCloud(width = 1500, height = 1000, stopwords= 불용어, background_color= 'white')
wc.generate(text)
plt.imshow(wc)
plt.show()
```

```python
불용어  = STOPWORDS.union({'goals', 'field', 'human', 'see'})
wc = WordCloud(width = 1500, height = 1000, stopwords= 불용어, background_color= 'white')
wc.generate(text)
plt.imshow(wc)
plt.axis('off') # 단위 없애주는 코드
plt.show()
```

```python
불용어  = STOPWORDS.union({'goals', 'field', 'human', 'see'})
maskHeart = np.array(Image.open('/image.png'))
wc = WordCloud(width = 1500, height = 1000, stopwords= 불용어, background_color= 'white', mask = maskHeart)
wc.generate(text)
plt.imshow(wc)
plt.axis('off') # 단위 없애주는 코드
plt.show()
```

```python
불용어  = STOPWORDS.union({'goals', 'field', 'human', 'see'})
maskHeart = np.array(Image.open('/image.png'))
wc = WordCloud(width = 1500, height = 1000, stopwords= 불용어, background_color= 'white', mask = maskHeart)
wc.generate(text)
plt.imshow(wc)
plt.axis('off') # 단위 없애주는 코드
plt.savefig('Aiwordcloud.png') # 사진 폴더에 저장해주는 코드
plt.show()
```

## 정규식 : 문자열 패턴을 나타내는 식 - 문자열을 처리하고 분석하는데 매우 유용
          : 특정 패턴의 문자열을 검색 / 추출 - 다양한 문자열 처리 작업 쉽게 수행
          

```python
import re # 정규식 라이브러리
text = "이번 학기 개설 과목은 CS101입니다."
match = re.search('[A-Za-z]{2}\d{3}', text)
if match :
  print("과목 코드가 발견됨 : " , match)
else:
  print("과목 코드가 발견되지 않음 : ", match)
```

```python
string = "My phone number is 010-1234-5678. call me anytime"
result = re.search('\d{3}-\d{4}-\d{4}', string)
print(result.group())
```

```python
text = 'There are 12 apples and 3 oranges '
numbers = re.findall('\d+', text)
print(numbers)
```

```python
def is_valid_jumin(st):
  pattern = '^\d{6}\-\d{7}'
  match = re.match(pattern, st)
  return match

jumin = input('당신의 주민등록번호는 : ')
if is_valid_jumin(jumin):
  print('유효한 주민등록번호입니다.')
else:
  print("주민등록번호가 유효하지 않습니다.")
```
