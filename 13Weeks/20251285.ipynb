```python
# folium,Map(속성값) - 지도를 그려주는 객체
import folium
from folium.plugins import MarkerCluster
```

```python
# 동양미래대학교 위치를 지도에 Marker 표시하기
map = folium.Map(location= (37.5007, 126.8677) ,
                 zoom_start= 13,
                 zoom_control= False,
                 control_scale= True)
# folium.Marker(속성들).add_to(지도변수명) 마커 표시하기
mk = folium.Marker([37.5007, 126.8677], tooltip = '동양미래대학교',
                   popup = folium.Popup("동양미래대학교", max_width= 100))
mk.add_to(map)
mk1 = folium.Marker([37.4992, 126.8918], tooltip = '거리공원')
mk1.add_to(map)
mk2 = folium.Marker([37.4887, 126.8354], popup = '항동철길',
                    icon = folium.Icon())
mk2.add_to(map)
mk3 = folium.Marker([37.4839, 126.8255], popup = '푸른수목원',
                    icon = folium.Icon(color = 'red', icon = 'pen-nib', prefix = 'fa'))
mk3.add_to(map)
mk4 = folium.Circle([37.4980, 126.8670], popup = '고척돔',
                    color = 'green')
mk4.add_to(map)
mk5 = folium.Circle([37.5003, 126.8369], popup="온수도시자연공원",
                    color = 'green', radius = 100)
mk5.add_to(map)
mk6 = folium.Circle([37.4919, 126.8440], popup = "평강성서유물박물관",
                     color = 'red', radius = 300, fill = True)
mk6.add_to(map)

mk5 = folium.Circle([37.5010596,126.8826803] , popup = "NC 백화점 신구로점",
                    color = 'green', radius = 500, fill = True)
mk5.add_to(map)
map
```

```python
mcVar = MarkerCluster().add_to(map)
folium.Marker([37.4009, 126.8695]).add_to(mcVar)
folium.Marker([37.4019, 126.8695]).add_to(mcVar)
folium.Marker([37.4029, 126.8695]).add_to(mcVar)
folium.Marker([37.4039, 126.8695]).add_to(mcVar)
folium.Marker([37.4049, 126.8695]).add_to(mcVar)
folium.Marker([37.4059, 126.8695]).add_to(mcVar)
map

map.save('map.html')
```
